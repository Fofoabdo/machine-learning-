```python
import numpy as np 
import pandas as pd
```

```python
data=pd.read_csv('Ecommerce Customers')
data
```

```python
df=data['Address']
```

```python
data
```

```python
data1=data['Address']
df = pd.DataFrame(data1)

# define a lambda function to split the address string into street, city/state, and zip
split_address = lambda x: pd.Series([x[0].strip(), ' '.join(x[1:-1]).strip(), x[-1].strip()])

# apply the lambda function to the address column to create new columns for street, city/state, and zip
df[['street', 'city_state', 'zip']] = df['Address'].str.split().apply(split_address)

# split the city_state column into city and state
df[['city', 'state']] = df['city_state'].str.split(',', expand=True)

# drop the original address and city_state columns

df=pd.DataFrame(df)

df
```

```python
data['street']=df['street']
data['city_state']=df['city_state']
data['zip']=df['zip']
data['city']=df['city']
data['state']=df['state']
```

```python
data
```

```python
# create a sample DataFrame with an email column
data2=data['Email']
df = pd.DataFrame(data2)

# use the str.split() method to split the email column into two parts: the name and the domain
df[['name', 'domain']] = df['Email'].str.split('@', expand=True)

# use string slicing to extract the part of the domain that represents the email type
df['type'] = df['domain'].str.split('.', expand=True)[0]

# drop the original email and domain columns
df = df.drop(['Email', 'domain'], axis=1)
```

```python
df2=pd.DataFrame(df)
df2
```

```python
data['name']=df2['name']
data['type']=df2['type']
```
