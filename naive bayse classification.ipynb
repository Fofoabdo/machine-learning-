```python
x=np.array(data.drop('doornumber',axis=1)).reshape(-1,1)
y=data['doornumber']
u=np.array(data['price']).reshape(-1,1)
```

```python
onehot=OneHotEncoder()
x=onehot.fit_transform(x)
```

```python
scaler=StandardScaler(with_mean=False)
x=scaler.fit_transform(x)
```

```python
selector=SelectKBest(f_classif,k=4)
x=selector.fit_transform(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
naive_bayes = MultinomialNB()

# Train the model
naive_bayes.fit(x_train, y_train)

# Make predictions on the test set
y_pred = naive_bayes.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, roc_auc_score

# Compute predictions
y_pred = naive_bayes.predict(x_test)

# Compute evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred,pos_label='four')
recall = recall_score(y_test, y_pred,pos_label='four')
f1 = f1_score(y_test, y_pred,pos_label='four')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler

# Generate synthetic imbalanced dataset
x, y = make_classification(n_samples=1000, n_features=5, weights=[0.9, 0.1], random_state=42)

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Apply Random Oversampling to the training data
ros = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros.fit_resample(X_train, y_train)

# Train Naive Bayes model on the resampled data
gnb = GaussianNB()
gnb.fit(X_train_resampled, y_train_resampled)

# Predict on test data
y_pred = gnb.predict(X_test)

# Calculate accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```
